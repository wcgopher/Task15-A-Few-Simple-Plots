```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

mean = 0
variance = 1
stdev = math.sqrt(variance)
# x = np.linspace(mean - 3*stdev, mean + 3*stdev,100)
x = np.linspace(-25,25)
plt.plot(x,mlab.normpdf(x, mean, math.sqrt(1)),'b',label = 'Variance 1')
plt.plot(x,mlab.normpdf(x, mean, math.sqrt(10)),'r',label = 'Variance 10')
plt.plot(x,mlab.normpdf(x, mean, math.sqrt(100)),'g',label = 'Variance 100')
plt.plot(x,mlab.normpdf(x, mean, math.sqrt(1000)),'y',label = 'Variance 1000')
data = np.random.uniform(-10,10,5) # You are generating 1000 points between 0 and 1.
count, bins, ignored = plt.hist(data, 20, facecolor='cyan', label = 'Uniform Distribution') 
plt.title("Normal Distributions with One Uniform Disribution")
plt.legend()

plt.show()


# data = np.random.uniform(0,1,1000) # You are generating 1000 points between 0 and 1.
# count, bins, ignored = plt.hist(data, 20, facecolor='magenta') 

# plt.xlabel('X~U[0,1]')
# plt.ylabel('Count')
# plt.title("Uniform Distribution Histogram (Bin size 20)")
# plt.axis([0, 1, 0, 100]) # x_start, x_end, y_start, y_end
# plt.grid(True)

# plt.show(block = False)
```

```python
t = linspace(0, 2*math.pi, 400)
a = sin(t)
b = cos(t)
c = a + b

plt.plot(t, a, 'r') # plotting t, a separately 
plt.plot(t, b, 'b') # plotting t, b separately 
plt.plot(t, c, 'g') # plotting t, c separately 
plt.show()
```
